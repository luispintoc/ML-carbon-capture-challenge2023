```python
import pandas as pd
from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/train_df_with_labels.csv')
df = df.dropna(subset='Target')
```

```python
df.drop(columns=['Date', 'Target', 'Abs Target', 'window_start'], inplace=True)
```

```python
df['window_duration'].unique()
```

```python
df = df[df['window_duration']>0]
```

```python
y = df['window_duration']
x = df.drop(['window_duration'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=123)
```

```python
clf = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```
